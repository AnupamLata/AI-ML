#**EDA**

Exploratory Data Analysis:
Exploratory Data Analysis (EDA) is the process of examining, summarizing, and visualizing a dataset to understand its main characteristics before building statistical models or machine learning algorithms.

**Main goals of EDA:**

1. *Understand the data structure*
* Number of rows and columns

* Data types (numerical, categorical, date, etc.)
* Variable names and meanings
2. *Clean the data*
* Identify missing values
* Detect duplicate records
* Correct inconsistent or incorrect entries
3. *Summarize the data*
* Calculate descriptive statistics:
  *  Mean
  * Median
  * Mode
  * Standard deviation
  * Minimum and maximum
  * Quartiles
*4. Visualize the data*
Common plots include:
* Histogram – shows the distribution of a numerical variable.
* Box plot – identifies outliers and spread.
* Bar chart – compares categories.
* Scatter plot – shows relationships between two numerical variables.
* Heatmap – displays correlations between variables.
5. Identify patterns and relationships
   * Trends
   * Correlations
   * Clusters
   * Seasonal effects (for time-series data)

***Common EDA tools***

For Python, the most commonly used libraries are:

  * Pandas – data manipulation and summary statistics.
  * NumPy – numerical computations.
  * Matplotlib – basic visualizations.
  * Seaborn – advanced statistical visualizations.

#**PANDAS**

Pandas is an open-source Python library used for data manipulation, analysis, and cleaning. It provides fast and easy-to-use data structures for working with structured data such as tables, spreadsheets, and CSV files.

**Main Data Structures**

**1. Series**

   * A one-dimensional labeled array.
   * Similar to a single column in a table.

```python
import pandas as pd

s = pd.Series([10,20,30,40])

print(s)
print(type(s))

#index


print(s[0])
print(s[2])
```

**Properties of Series**

```python
import pandas as pd

s = pd.Series([24,25,26,27,"abc"], index = ["Anupam","Riya", "Rahul","Bob", 25])

print(s)
```

2. Vectorized Operations

```python
import pandas as pd
s1 = pd.Series([1,2,3,4])
s2 = pd.Series([10,20,30,40])

print(s1+s2)
```

4. Mutable Values , immutable size

```python
import pandas as pd

s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([10,20,30,40,50])

s1[0] = 100

changed_s1 = s1.drop(1)

print(s1)
print(changed_s1)

```

**2. DataFrame**

---


   * A two-dimensional table with rows and columns.
   * Similar to an Excel spreadsheet or SQL table.

```python
#example
import pandas as pd

data ={
    "Name" : ["Anupam", "Riya", "Rahul"],
    "age" : [20,21,19]
}

df = pd.DataFrame(data)
print(df)
```

**Create DataFrame with dictionaries of lists :**

```python
# Example
import pandas as pd

info = {
    "Name" : ["Anupam", "Bob", "Adam"],
    "Age" : [23,24,25],
    "GPA" : [9.5, 8.5, 7.6]

}
df = pd.DataFrame(info)

print(df)
print(type(df))
print(df.index)  #for access row values
print(df.columns)
print(df.shape)
print(df.size)
```

**Create DataFrame with  a list of Lists:**

```python
import pandas as pd

data = [
    ["Aman", 20, "Delhi"],
    ["Ravi", 21, "Lucknow"],
    ["Priya" , 19, "Noida"]
]

df = pd.DataFrame(data, columns = ["Name", "Age", "City"])

print(df)
```

Create DataFrame with a numpy Arrays

**DataFrame Methods :**


---


A DataFrame method is a function used to perform operations on a DataFrame , such as viewing , cleaning , filtering, sorting, and analyzing data.

1. head() :  Displays the first 5 rows(default).

```python
import pandas as pd
df = pd.read_csv("employee_data.csv")

print(df.head())
print(df.head(3))
```

2. tail() :  Displays the last 5 rows.

```python
import pandas as pd

df = pd.read_csv("/content/employee_data.csv")

print(df.tail())
print(df.tail(2))
```

3. df.sample() :  The sample() method is used to randomly select rows or columns from a DataFrame.

```python
import pandas as pd

df = pd.read_csv("employee_data.csv")

print(df.sample(2))
```
