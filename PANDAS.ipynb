#**EDA**

Exploratory Data Analysis:
Exploratory Data Analysis (EDA) is the process of examining, summarizing, and visualizing a dataset to understand its main characteristics before building statistical models or machine learning algorithms.

**Main goals of EDA:**

1. *Understand the data structure*
* Number of rows and columns

* Data types (numerical, categorical, date, etc.)
* Variable names and meanings
2. *Clean the data*
* Identify missing values
* Detect duplicate records
* Correct inconsistent or incorrect entries
3. *Summarize the data*
* Calculate descriptive statistics:
  *  Mean
  * Median
  * Mode
  * Standard deviation
  * Minimum and maximum
  * Quartiles
*4. Visualize the data*
Common plots include:
* Histogram – shows the distribution of a numerical variable.
* Box plot – identifies outliers and spread.
* Bar chart – compares categories.
* Scatter plot – shows relationships between two numerical variables.
* Heatmap – displays correlations between variables.
5. Identify patterns and relationships
   * Trends
   * Correlations
   * Clusters
   * Seasonal effects (for time-series data)

***Common EDA tools***

For Python, the most commonly used libraries are:

  * Pandas – data manipulation and summary statistics.
  * NumPy – numerical computations.
  * Matplotlib – basic visualizations.
  * Seaborn – advanced statistical visualizations.

***1. Handle missing values:***

---



**isnull().sum()**:

It counts the number of missing (NaN) values in every column of a dataFrame.

```python
import pandas as pd

df = pd.read_csv("/content/raw_data.csv")

print(df.isnull().sum())
```

**dropna():**

The dropna() method is used to remove rows or columns that contain missing values(NaN).

```python
import pandas as pd

df = pd.read_csv("/content/raw_data.csv")

print(df.dropna())

# if drop the column:

print(df.dropna(axis = 1))
```

**fillna(value):**

fillna(value) fills all missing (NaN) values with the value you provide.

```python
import pandas as pd

df = pd.read_csv("/content/raw_data.csv")

# print(df.fillna(1))

age_mean = df["age"].mean()
print(df["age"].fillna(age_mean))

cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)
print(cleaned_data)

print(df)
```

**ffill() and bfill() :**

ffill() and bfill() are methods used to fill missing (NaN) values in a DataFrame or Series.
  * ffill() = Forward fill : fills a missing value with the previous (above) non-missing value.
  * bfill() = Backward fill : fills a missing value with the next (below) non-missing value.



```python
import pandas as pd

df = pd.read_csv("/content/raw_data.csv")

print(df)
print(df.ffill())

print(df.bfill())
```
